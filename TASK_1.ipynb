```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (skip metadata/header rows)
df = pd.read_csv('/content/API_SP.POP.TOTL_DS2_en_csv_v2_85220.csv', skiprows=4)

# Choose year to visualize (e.g., 2022)
year = '2022'

# Filter out rows with no data for that year
df_year = df[['Country Name', year]].dropna()

# Sort and get top 20 most populous countries for clarity
df_top = df_year.sort_values(by=year, ascending=False).head(20)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=year, y='Country Name', data=df_top, palette='viridis')
plt.title(f'Top 20 Most Populous Countries in {year}')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
